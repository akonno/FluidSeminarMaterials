# 3次元のグラフを描く―Matplotlib / Plotly

これまで2次元のグラフを描く練習をしたが、今回は3次元のグラフを描くことを試みる。「Matplotlibでグラフを描く」と「Numpy入門（超初歩のみ）」の教材をまだ学習していない場合は，先に学習すること．

グラフに表示する対象のデータとして、らせんのデータを作る。すなわち$(x,y)$方向には円運動をしていて、$z$方向にはだんだん大きくなるようなデータである。これを**Numpy**の機能を用いて作ってみよう。

````python
import numpy as np

t = np.linspace(0, 20*np.pi, 500)
x = np.cos(t)
y = np.sin(t)
z = t
````

はじめの```t = np.linspace(0, 20*np.pi, 500)```では、0～$20\pi$の間を500分割した数列を作成している。この場合、データには$20\pi$も含まれる。```np.pi```が$\pi$（円周率、3.141592…）である。

````python
print(t)
print(len(t))
````

それ次の```x = np.cos(t)```は変数```t```の余弦（コサイン）を計算している。tが500個の数値の列なので、xには500個の値が入る。つまりtの値のコサインが代入される。

同様にyの値も計算され、これによって変数x, y, zがそれぞれ500個の数値の列として作られた。以降ではこれを3次元のグラフにする。

## いきなり蛇足：numpyのlinspaceについて

今回の教材の趣旨からは外れるが、numpyのlinspaceについて注意しておく。

数値だと分かりづらいが、データtの末尾が$20\pi$（≒62.83…）になっていることが分かる。Pythonのfor文で```for n in range(500):```のようにループを回した場合は、nは0～499の範囲で実行されるが、```np.linspace```には**末尾の数値も含まれる**ことに注意されたい。

その一方で、配列の大きさは500である。つまり```t = np.linspace(0, 20*np.pi, 500)```では、0～$20\pi$の間を等間隔に**499分割して**、その値の列を返している。500分割ではない。

linspaceは生成したいデータ数を指定して等間隔の数列を生成するが、もしデータの間隔のほうを指定したければarangeという関数を利用できる。

````python
t2 = np.arange(0, 100, 2)
print(t2)
print(len(t2))
````

上の例では0～100を間隔2で分割している。すなわち50分割している。arangeの場合は末尾の数値（この場合は100）が含まれていないことに注意せよ。このためデータ数は50である。

なお実数（浮動小数点数）を用いた場合は、数値誤差のため末尾の数値（実際には、それにとても近い数値）が含まれることがある。

## Matplotlibで3Dグラフを描く

はじめに、すでに学んでいるmatplotlibで3Dグラフを描いてみる。

matplotlibで3次元の（$(x, y, z)$がある）グラフを描くには、座標軸が3Dであることをmatplotlibに指示する。その上で$(x, y, z)$のデータを与える。

````python
import matplotlib.pyplot as plt

# 3dであることを伝える。
ax = plt.figure().add_subplot(projection='3d')

ax.plot(x, y, z)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
````

これで3次元のグラフを描画することができる。$(x, y, z)$がらせんのデータであることが分かるであろう。

## Plotlyで3Dグラフを描く

[Plotly](https://plotly.com/graphing-libraries/)はインタラクティブかつきれいな（出版物に使えるクオリティの）グラフを作成することを目的としたライブラリである。PlotlyはPython専用ではなくR、Julia、JavaScriptなどからも使うことができる。matplotlibと同様、多数のグラフ描画機能を持ち美しいグラフを作成することができるパッケージだが、今回はPlotlyの簡易グラフ機能であるplotly.expressを利用してインタラクティブなグラフを作成する方法のみを紹介する。

はじめにPlotlyと，Plotlyから呼び出されるPandasをインストールする．

````python
!conda install -y plotly pandas
````

plotlyを新規にインストールしたあとは，matplotlibを新規にインストールしたときと同じく，ノートブックのメニューのKernel → Restart Kernel を実行せよ．

その後，JupyterLabノートブックに表示するため，「レンダラー」を設定する．（Google Colabを使っている場合は，この設定は不要）

````python
import plotly.io as pio
pio.renderers.default='notebook_connected'
````

ここまで設定すれば，グラフを描くためにやることはmatplotlibのときと大差ない。`plotly.express`をインポートして、描画したいデータを与えるだけである。

````python
import plotly.express as px

fig = px.line_3d(x=x, y=y, z=z)
fig.show()
````

matplotlibとの違いは、plotlyを用いるとJupyterLabであってもインタラクティブなグラフが表示されることである。グラフが表示されたら、マウス等で操作してみよ。視点を変えたり、拡大・縮小したりできるであろう。

なお拡大・縮小はマウスのホイールを用いれば容易に可能だが、ホイールが無いマウスを用いている場合は、右上に表示されるコントロール機能を使うことで可能である。（コントロールはマウスを近づけたときだけ表示される．）
