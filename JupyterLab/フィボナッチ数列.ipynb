# フィボナッチ数列（Fibonacci sequence）

フィボナッチ数列は「2つ前の数字と1つ前の数字を足すと、次の数字が出来る」というルールで作られる数の並びである．

最初の2つの数字として0と1が使われる．以降は2つ前の数字と1つ前の数字を足したものを並べていく．

0, 1, 1, 2, 3, 5, 8, 13, 21, ...

これをプログラムで書いてみよう．

```python
a = 0
b = 1
print(a, end=', ')
while b < 1000:
  print(b, end=', ')
  c = a + b
  a = b
  b = c
print()
```

フィボナッチ数列は自然の中でよく見られる形やパターンに関連していることが知られている．例えばひまわりの種の配置や，うずまきの形など，自然界にはこの数列が関連する美しいパターンがたくさん存在する．
